```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
movie_ratings=pd.read_csv('Desktop/Movie-Ratings.csv')
```

```python
movie_ratings.head()
```

### Q1)  list of top 10 movie(descending order) in terms of budget

```python
movie_ratings.sort_values('Budget (million $)',ascending=False).head(10)
```

###  Q2) list of top 10 movie (ascending order) in terms of budget

```python
movie_ratings.sort_values('Budget (million $)',axis=0,ascending=True).head(10)
```

### 3) which genre is the best in terms of price budget

```python
movie_ratings.groupby("Genre").mean().sort_values('Budget (million $)',ascending=True)
```
